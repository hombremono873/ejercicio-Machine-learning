# Función Open()

## Cómo funciona open()?
Abrir un archivo : El propósito principal de open()es abrir un archivo para interactuar con su contenido.
Devuelve un objeto de archivo : Este objeto te permite realizar operaciones específicas sobre el archivo.
Debe cerrarse después de usar : Puedes usar close()para liberar recursos, aunque con with esto se hace automáticamente.

## Sintaxis general
open(nombre_archivo, modo, encoding)

Parámetros principales :
nombre_archivo:

Es la
Ejemplo: "mi_archivo.txt","/carpeta/archivo.log".
modo(opcional):

Especifica lo que deseas hacer con el archivo. Los valores más comunes son:
"r":Leer (por
"w":Escribir. Sollozo
"a":
"x":Crear
"b":
"t":Texto
encoding:

Especifica la codificación del archivo (como utf-8).

# Formato csv

El formato CSV (valores separados por comas)
¿Qué es un archivo CSV?
CSV (valores separados por comas) es un,) Tama;) el tabú\t).

### Características principales:
Formato ligero y sencillo:
Es un archivo de texto sin formato, por lo que es fácil de leer y escribir tanto por humanos como por máquinas.

### Importancia del formato CSV:
Facilidad de intercambio de datos:

Los archivos CSV son una forma estándar de compartir datos entre diferentes sistemas, como bases de datos, hojas de cálculo y aplicaciones web.
Compatibilidad universal:

Prácticamente cualquier aplicación que maneje datos puede leer o exportar en formato CSV, lo que lo hace ideal para la integración de sistemas.
Ligereza:

Como son archivos de texto sin formato, su tamaño es pequeño en comparación con otros formatos como Excel ( .xlsx).
Facilidad de manipulación:

Se pueden abrir y editar fácilmente con un editor de texto, hojas de cálculo (como Excel), o mediante programación con bibliotecas especializadas.
Legibilidad humana:

Aunque está diseñado para máquinas, los archivos CSV son lo suficientemente simples para que los humanos los lean y editen directamente.

# Contexto del Módulo os

## Propósito General

Proveer herramientas para realizar operaciones del sistema operativo, como trabajar con archivos y directorios, manipular rutas, variables de entorno, procesos, permisos y más.
Funciona como un puente entre el código Python y las capacidades nativas del sistema operativo.
Portabilidad

Aunque el módulo funciona en todos los sistemas operativos soportados por Python, tiene diferencias específicas según el sistema en uso. Por ejemplo:
En Windows, os.name devuelve 'nt'.
En Linux/macOS, devuelve 'posix'.
Esto lo hace útil para escribir scripts portables que se adapten al entorno donde se ejecuten.
Áreas Clave del Módulo os

Gestión de Directorios: Cambiar el directorio actual, listar contenido, crear o eliminar carpetas.
Gestión de Archivos: Renombrar, mover, eliminar archivos, verificar su existencia.
Rutas de Archivos: Trabaja con rutas a través del submódulo os.path, que permite unir rutas, verificar tipos (archivo o carpeta), obtener rutas absolutas, etc.

Variables de Entorno: Permite acceder, modificar o agregar variables del entorno del sistema.
Ejecución de Comandos: Ejecutar comandos del sistema operativo directamente desde Python.
Manejo de Procesos: Obtener el ID del proceso actual, iniciar otros procesos o finalizar procesos existentes.

Creación de scripts para respaldos.
Limpieza de directorios temporales.
Modificación de permisos de archivos.
Ejecución de herramientas o scripts externos desde Python.

# Sugerencia

En este punto es requerido construir una función que permita leer el conjunto de datos de manera más limpia. Este tipo de prácticas son de gran utilidad para que el código sea más modular.

```python
def load_kdd_dataset(data_path):
    with open(data_path, 'r') as grupo_entreno:
        dataset=arff.load(grupo_entreno)
        attributes = [attr[0] for attr in dataset["attributes"]]
        return  pd.DataFrame(dataset["data"], columns = attributes)
        
```

```python
load_kdd_dataset("datasets/NSL-KDD/KDDTrain+.arff")
```

# LabelEncoder
LabelEncoder es una clase de preprocesamiento que nos permite codificar una caracteristica categorica a 0 o 1

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
df["class"] = labelencoder.fit_transform(df["class"])
categoricos = df.select_dtypes(include=['object', 'category']).columns.tolist()
df_encoded = pd.get_dummies(df, columns= categoricos)
df = df_encoded

```

# Buscamos correlaciones

Buscamos que atributos tienen correlacion con la etiqueta

por lo que me dicen que son fundamentales para la preiccion
