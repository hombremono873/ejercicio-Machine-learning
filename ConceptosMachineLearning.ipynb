# Overfitting
Si el conjunto de datos tiene muchas caracteristicas, es posible que la funcin de hiptesis generada por el algoritmo
se adapte muy bien al conjunto de datos de entrenamiento pero falle al generalizar con nuevos ejemplos, es lo que se conoce como overfitting

Significa que el algoritmo despues de ser entrenado y puesto a produccin, al enfrentar datos que no conoce falla.

El sobreajuste y el subajuste son dos de las principales razones por las que los algoritmos y modelos de aprendizaje automtico no obtienen buenos resultados. Comprender por qu surgen en primer lugar y tomar medidas para evitarlos puede mejorar el rendimiento de su modelo en muchos niveles. Exploremos mejor la diferencia entre sobreajuste y subajuste a travs de un ejemplo hipottico.

Supongamos que dos nios deben realizar un examen de matemticas. Debido a las limitaciones de tiempo, el primer nio solo aprendi a sumar y no pudo aprender a restar, multiplicar ni dividir. El segundo nio tena una memoria fenomenal, pero no era muy bueno en matemticas, por lo que, en lugar de eso, memoriz todos los problemas del libro de problemas. Durante el examen, el primer nio resolvi solo problemas matemticos relacionados con la suma y no pudo abordar problemas matemticos que involucraran las otras tres operaciones aritmticas bsicas. Por otro lado, el segundo nio solo fue capaz de resolver problemas que memoriz del libro de problemas de matemticas y no pudo responder ninguna otra pregunta. En este caso, si las preguntas del examen de matemticas fueran de otro libro de texto e incluyeran preguntas relacionadas con todo tipo de operaciones aritmticas bsicas, ambos nios no lograran aprobarlo.

Los algoritmos de aprendizaje automtico a veces muestran un comportamiento similar al de estos dos nios. Hay veces en las que aprenden solo de una pequea parte del conjunto de datos de entrenamiento (similar al nio que solo aprendi a sumar). En esos casos, el modelo no se ajusta lo suficiente. En otros casos, los modelos de aprendizaje automtico memorizan todo el conjunto de datos de entrenamiento (como el segundo nio) y funcionan de manera excelente en instancias conocidas, pero fallan en datos no vistos. En situaciones como esta, el modelo se ajusta demasiado. El ajuste excesivo y insuficiente son dos conceptos esenciales en el aprendizaje automtico y ambos pueden conducir a un rendimiento deficiente del modelo.


## Qu es la generalizacin?
En el aprendizaje supervisado , el objetivo principal es utilizar los datos de entrenamiento para crear un modelo que pueda hacer predicciones precisas basadas en datos nuevos e inditos, que tengan las mismas caractersticas que el conjunto de entrenamiento inicial. Esto se conoce como generalizacin. La generalizacin se relaciona con la eficacia con la que los conceptos aprendidos por un modelo de aprendizaje automtico se aplican a ejemplos particulares que no se utilizaron durante el entrenamiento. Se desea crear un modelo que pueda generalizar con la mayor precisin posible. Sin embargo, en el mundo real, este es un problema complicado.

Error de generalizacin
Como ya sabes, para entrenar un modelo de aprendizaje automtico, divides el conjunto de datos en tres conjuntos: entrenamiento, validacin y prueba. Como indican estos nombres, entrenas tus modelos usando los datos de entrenamiento, luego los comparas y los ajustas usando los resultados de la evaluacin en el conjunto de validacin y, por ltimo, evalas el rendimiento de tu mejor modelo en el conjunto de prueba. La tasa de error en los casos nuevos se denomina error de generalizacin (o error fuera de la muestra) y, al evaluar tus modelos en el conjunto de validacin, obtienes una estimacin de este error. Este valor te indica qu tan bien funcionan tus modelos en instancias en las que nunca se han iterado.

El error de generalizacin de un modelo (tambin conocido como error de prediccin) se puede expresar como la suma de tres errores muy diferentes: error de sesgo, error de varianza y error irreducible.

Nota: El error irreducible se produce debido al ruido de los datos en s. La nica forma de reducir esta parte del error es limpiar los datos (por ejemplo, reparar las fuentes de datos, como sensores daados, o detectar y eliminar valores atpicos).

## El concepto de sesgo: Error de sesgo

Este tipo de error es el resultado de suposiciones incorrectas, como pensar que los datos son lineales cuando en realidad son cuadrticos. El sesgo se define como un error sistemtico que ocurre en el modelo de aprendizaje automtico como resultado de suposiciones errneas de ML. El sesgo tambin es la diferencia cuadrtica promedio entre las predicciones del modelo y los datos reales. Los modelos con un mayor porcentaje de sesgo no coincidirn con los datos de entrenamiento. Por otro lado, los modelos con tasas de sesgo ms bajas coincidirn con el conjunto de datos de entrenamiento. Las caractersticas de un modelo con alto sesgo incluyen:

No se logra capturar tendencias de datos adecuadas
Potencial de subadaptacin
Ms generalizado/demasiado simplificado
Alta tasa de error

## El concepto de varianza: Error de varianza

La varianza, como error de generalizacin, se produce debido a la excesiva sensibilidad del modelo a pequeas variaciones en los datos de entrenamiento. En el aprendizaje supervisado, el modelo aprende de los datos de entrenamiento. Por lo tanto, si cambia los datos de entrenamiento, el modelo tambin se ver afectado. La varianza muestra la cantidad en la que se ver afectado el rendimiento del modelo predictivo al evaluar en funcin de los datos de validacin. Si su modelo puede generalizar bien, no debera cambiar demasiado de una divisin a otra. Por ejemplo, es probable que un modelo con muchos grados de libertad (como un modelo polinomial de alto grado) tenga una varianza alta, mientras que los modelos lineales probablemente tengan una varianza menor. Un modelo de alta varianza normalmente tiene las siguientes cualidades:

Ruido en el conjunto de datos
Potencial de sobreajuste
Modelos complejos
Intentando poner todos los puntos de datos lo ms cerca posible

## Compensacin entre sesgo y varianza
El sesgo/varianza en el aprendizaje automtico se relaciona con el problema de minimizar simultneamente dos fuentes de error (error de sesgo y error de varianza).





# Deteccin del overfitting

##  Cmo se detecta el overfitting?
Se puede detectar usando curvas de aprendizaje que muestran cmo evoluciona el error durante el entrenamiento.

Si el error en los datos de entrenamiento disminuye, pero el error en los datos de validacin aumenta, es probable que haya overfitting.


##  Cmo prevenir el overfitting?

### Aumentar la cantidad de datos de entrenamiento.
Un mayor volumen de datos ayuda al modelo a aprender patrones ms generales.

### Usar regularizacin.
La regularizacin agrega una penalizacin a los pesos grandes del modelo, lo que evita que se ajuste demasiado a los datos de entrenamiento.

1. L1 Regularization (Lasso): Penaliza la suma de los valores absolutos de los pesos.
2. L2 Regularization (Ridge): Penaliza la suma de los cuadrados de los pesos.
4. Reducir la complejidad del modelo.
   Por ejemplo, usar menos capas en una red neuronal o menos profundidad en un rbol de decisin.

5. Aplicar tcnicas de Early Stopping.
   Detener el entrenamiento cuando el rendimiento en los datos de validacin deja de mejorar.

6. Utilizar Cross-Validation.
   La validacin cruzada evala el modelo en diferentes subconjuntos del conjunto de datos para garantizar que generaliza bien.

7. Dropout (en redes neuronales).
   Durante el entrenamiento, aleatoriamente se apagan algunas neuronas para evitar que el modelo dependa demasiado de ciertas conexiones.

8. Eliminar caractersticas irrelevantes.
   Utilizar tcnicas de seleccin de caractersticas para evitar que el modelo aprenda patrones intiles.

#  Por Qu Dividir los Datos en Tres Grupos?

##  2. Datos de Validacin (Validation Set)
Este conjunto se usa para ajustar los hiperparmetros del modelo y seleccionar el modelo final. No se usa para entrenar, sino para evaluar cmo generaliza el modelo despus del entrenamiento.

###  Objetivo:
Evitar el sesgo hacia el conjunto de prueba al ajustar los hiperparmetros. Se usa para comparar modelos y seleccionar el mejor.

##  3. Datos de Prueba (Test Set)
Este conjunto se usa solo una vez al final del proceso, para evaluar el rendimiento del modelo final. Es importante que el modelo nunca haya visto estos datos durante el entrenamiento o la validacin.

###  Objetivo:
Obtener una estimacin realista del rendimiento del modelo en datos nuevos.

###  Qu Se Busca al Separar los Datos en Tres Conjuntos?
Evitar Overfitting:
Al usar datos de validacin y prueba, aseguras que el modelo no se sobreajuste al conjunto de entrenamiento.

### Evaluar la Generalizacin:
Separar los datos permite evaluar cmo el modelo generaliza a datos nuevos.

### Comparar Modelos de Manera Justa:
El conjunto de validacin permite una comparacin justa entre diferentes modelos y configuraciones.

### Ajustar Hiperparmetros sin Sesgo:
El conjunto de validacin permite ajustar los hiperparmetros sin sesgarse hacia el conjunto de prueba.

# Matriz de Confusin.


## Qu es la matriz de confucin ?
Es una herramienta fundamental en el anlisis de modelos de clasificacin, especialmente en aprendizaje automtico. Es una tabla que se utiliza para evaluar el rendimiento de un modelo de clasificacin al comparar las predicciones del modelo con los valores reales.

## Pero que es ?
Una matriz de confusin es una tabla de contingencia de tamao 2X2 (Para clasificacion binaria) o n X n (Para clasificacin multiclase), que resume los resultados de las predicciones de un modelo de clasificacin en trminos de:
- Predicciones correctas: El modelo predijo correctamente la clase.
- Predicciones incorrectas: El modelo predijo correctamente la clase.
  

- TP (Verdaderos positivos): Son los casos en los que el modelo predijo correctamente que una observacin pertenece a la clase positiva.
- FN (Falsos negativos) : Son los casos en los que el modelo predijo incorrectamente que una observacin pertenece a la clase negativa.
- FP (Falsos positivos) : Son los casos en los que el modelo predijo incorrectamente la clase positiva.
- TN (Verdaderos negativos) : Son los casos en los que el modelo predijo correctamente la clase negativa.

# Problema que resuelve la matriz de confusin.

La matriz de confusin ayuda a entender no solo la precisin global del modelo, sino tambin qu tipos de herrores est cometiendo. Mientras que las mtricas comunes como la precisin (accuracy) puede dar una idea general de como se desempea el modelo, la matriz de confusin proporciona un anlisis ms detallado sobre los tipos de errores cometidos.

Esto es til especialmente en escenarios desbalanceados, donde el modelo puede predecir bien la clase mayoritaria (por ejemplo los negativos), pero fallar al predecir correctamente la clase minoritaria (Los positivos).
La matriz de confusin nos ayuda a identificar esos casos y a mejorar el modelo.
