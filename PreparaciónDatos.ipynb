# Preparacin de datos 



La preparacin de los datos es una fase crucial en el proceso de desarrollo de un modelo de Machine Learning. Consiste en limpiar, transformar y estructurar los datos para que puedan ser utilizados de manera efectiva en el entrenamiento del modelo. Esta etapa impacta directamente en el rendimiento del modelo, ya que los datos deben ser de alta calidad y adecuados para el tipo de algoritmo que se va a utilizar.

## Pasos en preparacin de los datos

## 1. Recoleccin de Datos:
La preparacin comienza con la obtencin de datos. Los datos pueden provenir de diferentes fuentes como archivos CSV, bases de datos, API, encuestas, etc.

Ejemplo: Descargar un archivo CSV con datos de ventas de productos o acceder a una base de datos de transacciones bancarias.

## 2. Limpieza de Datos:
Esta etapa implica eliminar o corregir datos errneos, incompletos o irrelevantes para asegurar que los datos sean consistentes.

### Tareas comunes:

Manejo de valores faltantes: Los valores nulos o vacos deben ser manejados (puedes eliminarlos o imputarlos con un valor medio, mediano, etc.).
Eliminacin de duplicados: Asegurarse de que no haya registros repetidos en el conjunto de datos.
Deteccin de outliers: Identificar y tratar valores atpicos que pueden influir negativamente en el modelo.

## 3. Transformacin de Datos:
Aqu los datos se ajustan a un formato adecuado para el modelo. Esto puede incluir la normalizacin o escalado de caractersticas, conversin de tipos de datos y la creacin de nuevas caractersticas.

### Tareas comunes:

Escalado de caractersticas: Algunas tcnicas de Machine Learning (como SVM o redes neuronales) son sensibles a la escala de los datos, por lo que puede ser necesario escalar las caractersticas (por ejemplo, normalizando entre 0 y 1 o estandarizando a media 0 y desviacin estndar 1).
Codificacin de variables categricas: Convertir variables categricas en variables numricas, por ejemplo, utilizando One-Hot Encoding o Label Encoding.
Creacin de nuevas caractersticas: A veces es til generar nuevas variables a partir de las existentes, como transformar fechas en aos, meses o das.

# One hot encoding

El One-Hot Encoding es una tcnica de transformacin utilizada en Machine Learning para convertir variables categricas (aquellas que toman un nmero limitado de valores, como colores, tipos de producto, etc.) en variables numricas. Esta tcnica es esencial porque la mayora de los algoritmos de Machine Learning requieren que las caractersticas sean numricas.

Cmo funciona el One-Hot Encoding?
Supongamos que tienes una columna de datos categricos llamada "Color" con los siguientes valores: Rojo, Verde, y Azul. El objetivo del One-Hot Encoding es convertir estos valores en columnas binarias (0 o 1), donde cada columna representa una categora posible.



## En el ejemplo anterior:

Rojo se convierte en [1, 0, 0].
Verde se convierte en [0, 1, 0].
Azul se convierte en [0, 0, 1].
Cada fila original se convierte en un conjunto de nuevas columnas binarias, donde:

1 indica que la fila pertenece a esa categora.
0 indica que la fila no pertenece a esa categora.

## Beneficios de One-Hot Encoding:

Representacin Numrica: Convierte las categoras en nmeros que los algoritmos de Machine Learning pueden procesar.
Sin Implicar Orden: A diferencia de la codificacin de etiquetas (Label Encoding), One-Hot Encoding no implica un orden entre las categoras. Por ejemplo, en un conjunto de colores, no tiene sentido asignar el nmero 1 a "Rojo", 2 a "Verde" y 3 a "Azul", ya que no hay una relacin ordinal entre ellos.
Mejora del Rendimiento: En algunos algoritmos como rboles de decisin o redes neuronales, la tcnica de One-Hot Encoding puede mejorar la interpretacin y el rendimiento, ya que las categoras se representan como vectores binarios independientes.

Desventajas:
Aumento de la Dimensionalidad: Si tienes una columna con muchas categoras nicas, la matriz resultante puede ser muy grande y es posible que tengas problemas de alta dimensionalidad.
Sparsity (Escasez de datos): La representacin en la que la mayora de los valores son 0 puede causar que los datos sean dispersos, lo que puede ser ineficiente para almacenar y procesar.
Implementacin en Python (con Pandas):
Para realizar One-Hot Encoding en un DataFrame, puedes usar la funcin get_dummies de Pandas.

import pandas as pd

# Crear un DataFrame con una columna categrica
data = {'Color': ['Rojo', 'Verde', 'Azul', 'Rojo']}
df = pd.DataFrame(data)

# Aplicar One-Hot Encoding usando pd.get_dummies()
df_encoded = pd.get_dummies(df, columns=['Color'])

print(df_encoded)



```python
import pandas as pd
data = {
    'Color': ['Rojo', 'Verde', 'Azul', 'Rojo']
}
df = pd.DataFrame(data)
print(df)
df_encoded = pd.get_dummies(df, columns=['Color'])
df_encoded = df_encoded.astype(int)     #Convierte valores logicos al typo int
print(df_encoded)
```
