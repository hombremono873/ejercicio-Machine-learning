# Estudio de graficas de barras

## 1. ¿Qué es un gráfico de barras?
Un gráfico de barras es un tipo de gráfico que representa datos categóricos mediante barras rectangulares. La longitud de cada barra es proporcional al valor que representa. Los gráficos de barras son ideales cuando quieres comparar categorías entre sí.

Existen dos tipos principales de gráficos de barras:

Gráfico de barras verticales (barras hacia arriba o hacia abajo): Las barras se dibujan de manera vertical.
Gráfico de barras horizontales (barras hacia los lados): Las barras se dibujan de manera horizontal.
En Matplotlib, el gráfico de barras verticales se crea utilizando plt.bar(), mientras que el gráfico de barras horizontales se crea con plt.barh().

## 2. ¿Cuándo usar un gráfico de barras?
Los gráficos de barras son útiles cuando:

Comparas valores entre diferentes categorías: Por ejemplo, para comparar las ventas de diferentes productos o la cantidad de estudiantes en diferentes grados.
Tienes datos categóricos: Las categorías pueden ser etiquetas como nombres de productos, tipos de personas, etc.
Quieres destacar la magnitud de las diferencias: La longitud de las barras permite ver rápidamente cuál es la categoría más grande o más pequeña.

## 3. Estructura de un gráfico de barras
En un gráfico de barras típico:

Eje X (horizontal): Representa las categorías (por ejemplo, nombres de productos, meses del año, etc.).
Eje Y (vertical): Representa los valores numéricos (por ejemplo, cantidad de ventas, número de personas, etc.).

## 4. Componentes clave del gráfico de barras
Barras: Las barras son la representación visual de los datos. En un gráfico de barras verticales, la altura de la barra indica el valor de la categoría. En un gráfico de barras horizontales, la longitud de la barra indica el valor.

Eje X (Categorías): Las categorías o grupos que estás comparando se representan en el eje X. Pueden ser valores cualitativos o discretos.

Eje Y (Valores numéricos): El eje Y muestra los valores numéricos asociados con cada categoría en el eje X. Por ejemplo, podría mostrar cantidades, porcentajes o frecuencias.

Etiquetas: Se usan para identificar las categorías en el eje X y las unidades en el eje Y.

Leyenda (opcional): Si hay diferentes grupos representados por distintas barras, se puede incluir una leyenda para identificar cada grupo.

# 2. Gráfico Interactivo con Plotly
Para agregar interactividad como zoom, hover, y clics, puedes usar Plotly. Aquí tienes un ejemplo con un gráfico de barras interactivo:

```python

# Crear DataFrame
data = {'Producto': ['A', 'B', 'C', 'D', 'E'],
        'Ventas': [200, 350, 150, 400, 250]}
df = pd.DataFrame(data)

# Crear gráfico de barras interactivo
fig = px.bar(df, x='Producto', y='Ventas', 
             title='Ventas por Producto',
             labels={'Producto': 'Productos', 'Ventas': 'Ventas'},
             color='Ventas', color_continuous_scale='Blues')

# Personalizar aspecto
fig.update_layout(
    xaxis_title='Productos',
    yaxis_title='Ventas',
    template='plotly_dark',  # Estilo oscuro para más elegancia
    font=dict(family="Arial, sans-serif", size=14)
)

# Mostrar gráfico interactivo
fig.show()
```

# Pasos para crear un grafico básico en 3D con Matploblib
En python uno de los paquetes más comunes para crear gráficos es matplotlib, y para graficos 3D  se usa el modulo axes3D

## 1. fig = plt.figure()
Este código crea una nueva figura para tu gráfico. Pero, ¿qué es una figura en matplotlib?

### Figura (Figure)
En matplotlib, una figura es un contenedor general para uno o más gráficos o ejes. Es el espacio en el que se dibujan tus gráficos. Piensa en la figura como el lienzo en el que se dibujarán todos tus gráficos.

Ejemplo de figura: Una hoja de papel en blanco donde puedes dibujar gráficos.
Cuando usas fig = plt.figure(), estás creando una nueva figura vacía donde más adelante agregarás gráficos. Puedes tener varias figuras, cada una con diferentes gráficos.

Parámetros del método figure():
Si pasas parámetros adicionales, puedes personalizar el tamaño de la figura o los márgenes, entre otras cosas. Ejemplo:

figure = plt.figure(figsize=(12, 8))  # Crear figura con tamaño 8x6 pulgadas

### 2. ax = fig.add_subplot(111, projection='3d')
Este es un paso fundamental para trabajar con gráficos en matplotlib, y vamos a explicarlo a fondo.

Ejes (Axes)
En matplotlib, los ejes (también conocidos como Axes) son las áreas dentro de una figura donde se dibujan los gráficos. Cuando hablas de un gráfico en matplotlib, generalmente estás hablando de los ejes dentro de una figura.

## Método add_subplot()
El método add_subplot() permite añadir un conjunto de ejes a la figura. La sintaxis básica es:
### ax = fig.add_subplot(nrows, ncols, index)

## ¿Qué significa 111?
111 se desglosa de la siguiente forma:
1 fila de gráficos.
1 columna de gráficos.
1 es el índice del gráfico (solo hay uno, así que es el primero).
Por lo tanto, add_subplot(111) crea un único gráfico dentro de la figura, ocupando toda la figura (una sola celda de la cuadrícula).

### projection='3d'
Esto es lo que hace que los ejes (ax) sean 3D. Cuando usas projection='3d', estás indicando que los ejes deben ser capaces de manejar gráficos en tres dimensiones. Sin este parámetro, los ejes serían bidimensionales, es decir, gráficos planos.

### ¿Por qué usamos '3d'?:
Para poder crear gráficos en tres dimensiones, como gráficos de dispersión o de superficies, que necesitan una tercera dimensión para representar los datos.

```python
x = np.random.rand(100)  #100 datos aleatorios en el eje x
y = np.random.rand(100)  #100 datos aleatorios en el eje y
z = np.random.rand(100)  #100 datos aleatorios en el eje z
fig = plt.figure(figsize=(12,8))                     #Se crea el esopacio o contenedor
ejes = fig.add_subplot(111, projection='3d')           #Prepramos el contenedor para crear una figura
ejes.scatter(x,y,z)
ejes.set_xlabel("Eje X")
ejes.set_ylabel("Eje y")
ejes.set_zlabel("Eje z")
plt.title("Grafico de dispersion")
plt.show()
```

# Ejemplo 1: Superficie 3D de una Función Matemática
Una de las formas más comunes de utilizar gráficos 3D es representar una superficie que sigue una fórmula matemática. Vamos a crear una superficie 3D utilizando una función matemática simple, como 
𝑧=sin(𝑥2+𝑦2)
z=sin(x2+y2).

### Código para graficar la superficie 3D:

```python
#Creamos datos en una malla 2D

x = np.linspace(-5, 5 , 100)
y = np.linspace(-5,5,100)

x,y = np.meshgrid(x,y) #creamos una malla organizada de puntos (x,y) la cuadricula es 2D

z = np.sin(x**2 + y**2)

fig = plt.figure(figsize=(12,8))  #Contenedor de la imagen
ax = fig.add_subplot(111, projection="3d")  #Generamos el sistema de ejes con proyeccion espacial

#graficar  la superficie
ax.plot_surface(x,y,z, cmap='viridis')  #Imagen de superficie

ax.set_xlabel("Eje X")
ax.set_ylabel("Eje Y")
ax.set_zlabel("Eje Z")
plt.show()
```

# Malla de datos
Crear una malla en el contexto de gráficos 3D (y en general en matemáticas y computación) significa construir una estructura de puntos en un espacio multidimensional, de forma que cada punto está posicionado sobre una cuadrícula regular de valores. Este proceso es crucial cuando se quieren representar superficies o estructuras que dependen de dos o más variables (como las ecuaciones en 3D) de manera visual.
¿Por qué crear una malla?
Cuando trabajamos con gráficos 3D que involucran funciones de más de una variable (por ejemplo, 
𝑧=𝑓(𝑥,𝑦)
z=f(x,y)), necesitamos evaluar la función en diferentes combinaciones de los valores de 𝑥 x e 𝑦

## ¿Por qué es importante crear una malla?
La creación de una malla nos permite:

Evaluar funciones de múltiples variables: Muchas funciones matemáticas, especialmente las que representan superficies, dependen de más de una variable. La malla permite evaluar esas funciones en un espacio bidimensional (para funciones de 
𝑥
x e 
𝑦
y) o tridimensional.
Visualizar superficies complejas: Sin una malla, no tendríamos una manera sencilla de representar superficies en 3D, ya que no podríamos obtener suficientes puntos para trazar la forma completa.
Controlar la resolución: Cuantos más puntos definamos en nuestra malla (a través de np.linspace), más detallada será la superficie visualizada. A medida que incrementamos la cantidad de puntos en la malla, la superficie será más suave y precisa.


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Crear un conjunto de datos en el espacio angular (theta y phi)
theta = np.linspace(0, 2*np.pi, 100)  # Ángulo alrededor del eje z
phi = np.linspace(0, 2*np.pi, 100)    # Ángulo alrededor del toroide

# Crear una malla para theta y phi
theta, phi = np.meshgrid(theta, phi)

# Definir los parámetros del toroide
R = 5  # Radio mayor
a = 1  # Radio menor

# Ecuaciones paramétricas del toroide
x = (R + a * np.cos(phi)) * np.cos(theta)
y = (R + a * np.cos(phi)) * np.sin(theta)
z = a * np.sin(phi)

# Crear la figura
fig = plt.figure(figsize=(10, 8))

# Crear el conjunto de ejes 3D
ax = fig.add_subplot(111, projection='3d')

# Graficar la superficie
ax.plot_surface(x, y, z, cmap='inferno', edgecolor='none')

# Añadir etiquetas y título
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_zlabel('Eje Z')
ax.set_title('Toroide 3D')

# Mostrar el gráfico
plt.show()
```
