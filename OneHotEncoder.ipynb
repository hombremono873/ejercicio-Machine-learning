# One_Hot_Encoding



1. One hot Encoding: Crea una columna binaria, donde se asigna un uno si el registro pertenece a la categoría y 0 en caso contrario.
2. Uso:
   - Ideal para variables nominales:
   - Puede generar muchas columnas si hay muchas categorías
# Ejemplo

```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np

data = np.array(["Rojo", "Verde", "Azul", "Rojo", "Verde"]).reshape(-1,1)
encoder = OneHotEncoder(sparse_output=False)
encoded= encoder.fit_transform(data)
print(encoded)
```

# Nota sobre el uso del metodo reshape(-1,1)
Cuando se trabaja con la librería de scikit-learn muchos de sus métodos esperan una matriz bidimencional, pero si nuestro conjunto de datos es una array, como es el caso de una lista lista.reshape(-1,1)  transforma el array en uno bidimencional

## Ejemplo
colores = ["Rojo", "Verde", "Azul", "Rojo", "Verde"]

color_trans = colores.reshape(-1,1)

color_trans = [["Rojo"], ["Verde"], ["azul"]]
