# Transformadores variables categóricas


# Label Encoder
1. Descripción: Asigna un número único a cada categoría. Por ejemplo, para la variable color ["rojo", "verde", "azul2]
    rojo -> 0, verde -> 1 y azul -> 2
3. Uso: Adecuado para variables ordinales (Oden implícito, "bajo", "medio", "alto"].
   - No recomendado para variables no ordinales, ya que los modelos podrían interpretar los numeros como relaciones jerárquicas.
   - 
# Ejemplo

```python
from sklearn.preprocessing import LabelEncoder

color = ["Rojo", "Verde","Azul", "Rojo", "Verde"]
encoder = LabelEncoder()
color_codificado = encoder.fit_transform(color)

print(color_codificado)
```
